```python
def get_identical_values(grants_df, column_name):
    values = grants_df.groupby('queried_project_id')[column_name].agg('first')
    # Validation check for identical values
    if not values.nunique() == 1:
        raise ValueError(f"Validation failed: {column_name} does not have identical values for all rows with the same project_id.")
    return values

def get_newest_values(grants_df, column_name):
    # Find the index of the row with the maximum 'award_notice_date'
    idxmax_date = grants_df.groupby('queried_project_id')['award_notice_date'].idxmax()
    # Extract values based on the found index
    return grants_df.loc[idxmax_date, column_name]

def get_oldest_values(grants_df, column_name):
    # Find the index of the row with the minimum 'award_notice_date'
    idxmin_date = grants_df.groupby('queried_project_id')['award_notice_date'].idxmin()
    # Extract values based on the found index
    return grants_df.loc[idxmin_date, column_name]
```

```python
def gather_project_data(grants_df):
    # Create a new DataFrame for projects
    projects_df = pd.DataFrame()

    # Columns to aggregate with corresponding aggregation type
    aggregation_columns = {
        'project_title': 'newest',
        'abstract_text': 'newest',
        'project_start_date': 'oldest',
        'project_end_date': 'newest',
        'opportunity_number': 'identical',
        'api_source_search': 'identical',
        'org_name': 'identical',
        'org_city': 'identical',
        'org_state': 'identical',
        'org_country': 'identical',
        'program.program_id': 'identical'
    }

    # Apply aggregation for each column
    for column, aggregation_type in aggregation_columns.items():
        if column not in grants_df:
            raise ValueError(f"Expected column not found in grants data: {column}.")
        else:
            if aggregation_type == 'identical':
                projects_df[column] = get_identical_values(grants_df, column)
            elif aggregation_type == 'newest':
                projects_df[column] = get_newest_values(grants_df, column)
            elif aggregation_type == 'oldest':
                projects_df[column] = get_oldest_values(grants_df, column)
    
    return projects_df
```

```python
def validate_identical_values(df, field_name):
    inconsistent_rows = df[df.groupby('queried_project_id')[field_name].transform('nunique') > 1]
    for index, row in inconsistent_rows.iterrows():
        print(f"NOTICE: Unexpected mismatch in {field_name} for {row['queried_project_id']}")

def get_newest_or_oldest_value(df, field_name, newest=True):
    return df.sort_values('award_notice_date', ascending=not newest).groupby('queried_project_id')[field_name].first()

def gather_project_data(grants_df):
    """Aggregates grants data into project-level information, handling inconsistencies and validation."""

    field_aggregation_map = {
        'queried_project_id': 'Identical',
        'project_title': 'Newest',
        'abstract_text': 'Newest',
        'project_start_date': 'Oldest',
        'project_end_date': 'Newest',
        'opportunity_number': 'Identical',
        'api_source_search': 'Identical',
        'org_name': 'Identical',
        'org_city': 'Identical',
        'org_state': 'Identical',
        'org_country': 'Identical',
        'program.program_id': 'Identical'
    }

    projects_df = pd.DataFrame(columns=field_aggregation_map.keys())

    # Validate and aggregate "Identical" fields
    for field_name, aggregation_type in field_aggregation_map.items():
        if aggregation_type == 'Identical':
            validate_identical_values(grants_df, field_name)
            projects_df[field_name] = grants_df.groupby('queried_project_id')[field_name].first()

    # Get "Newest" or "Oldest" fields
    for field_name, aggregation_type in field_aggregation_map.items():
        if aggregation_type in ('Newest', 'Oldest'):
            projects_df[field_name] = get_newest_or_oldest_value(grants_df, field_name, newest=aggregation_type == 'Newest')

    # Rename 'queried_project_id' to 'project_id'
    projects_df.rename(columns={'queried_project_id': 'project_id'}, inplace=True)

    return projects_df
```

```python
def test_newest_oldest_project_val(grants_df, project_id):
    """Test that correct project values are pulled from either newest or oldest 
    grants as defined in config."""

    # Isolate a single project from input df
    df = grants_df[grants_df['queried_project_id'] == project_id]

    
    # Build empty projects df
    projects_df = pd.DataFrame(columns=field_aggregation_map.keys())
    
    # Set date column
    date_column = 'award_notice_date'

    # Get oldest and newest dates
    oldest = (df.sort_values(by=date_column, ascending=True)
              [date_column].unique().tolist()[0])
    newest = (df.sort_values(by=date_column, ascending=True)
                [date_column].unique().tolist()[-1])
    
    # If oldest or newest values are NaN, use fiscal year
    if pd.isna(oldest) or pd.isna(newest):
        
        date_column = 'fiscal_year'
        oldest = (df.sort_values(by=date_column, ascending=True)
              [date_column].unique().tolist()[0])
        newest = (df.sort_values(by=date_column, ascending=True)
                [date_column].unique().tolist()[-1])

    # Run project workflow to get actual value
    # Validate and aggregate fields based on aggregation type
    for field_name, aggregation_type in field_aggregation_map.items():
        if aggregation_type in ('newest', 'oldest'):
            projects_df[field_name] = get_newest_or_oldest_value(df, 
                                                                 field_name,
                                                                 #date_field=date_column,
                                                                 newest=aggregation_type == 'newest')
            
    # Define dict for testing 
    test_newest_oldest = {
        'project_start_date': oldest,
        'project_end_date': newest,
        'project_title': newest,
        'abstract_text': newest
    }

    # Build empty list
    valid_list = []

    try:

        # Loop through each column and compare actual vs expected values
        for col_name, date in test_newest_oldest.items():
            actual = projects_df[col_name].iloc[0] 
            expected = df[df[date_column] == date][col_name].iloc[0]
        
            if actual == expected:
                valid = True
            else:
                valid = False
                print(f"Mismatch in {col_name} for {project_id}:\n"
                    f"\tActual:   {actual}\n"
                    f"\tExpected: {expected}")
                
            # Add to running list
            valid_list.append(valid)

        # if all(valid_list):
        #     print(f"All good. Actual values match expected for {project_id}.")
    
    except Exception as error:
        print(f"Issue encountered with {project_id}: {error}\n"
              f"{display(df.head(2))}")
        
    return all(valid_list)
```

```python
all_results = list()
for project in grants_df['queried_project_id'].unique():
    project_results = test_newest_oldest_project_val(grants_df, project)
    all_results.append(project_results)

if all(all_results) == True: print("All good!")
```
